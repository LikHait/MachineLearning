```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

Считываем данные, удаляем первый столбец - нумерацию строк

```python
aggressive = ["Elbowing.txt", "Frontkicking.txt", "Hamering.txt", "Headering.txt", "Kneeing.txt", "Pulling.txt", "Punching.txt"]
normal = ["Bowing.txt", "Clapping.txt", "Handshaking.txt", "Hugging.txt", "Jumping.txt", "Running.txt", "Seating.txt"]
data = []
for file_name in aggressive:
    file = open('C://Users//Boris//Downloads//Vicon_Physical_Action_Data_Set//sub1//aggressive//' + file_name, 'r')
    tmp_data = pd.read_csv(file, sep=' ', skipinitialspace=True, header=None, skiprows=1)
    tmp_data["action"] = file_name.replace(".txt", "")
    tmp_data["type"] = 'agressive'
    data.append(tmp_data)
    
for file_name in normal:
    file = open('C://Users//Boris//Downloads//Vicon_Physical_Action_Data_Set//sub1//normal//' + file_name, 'r')
    tmp_data = pd.read_csv(file, sep=' ', skipinitialspace=True, header=None, skiprows=1)
    tmp_data["action"] = file_name.replace(".txt", "")
    tmp_data["type"] = 'normal'
    data.append(tmp_data)
    
data = pd.concat(base for base in data)
data = data.drop(data.columns[0], axis=1)
```

# Нормализация количественных признаков

```python
data_numerical = data[numerical_columns]
data_numerical.describe()
```

Нормализуем данные

data_numerical.mean - Среднее значение.

data_numerical.std - Стандартное отклонение наблюдений.

```python
data_numerical = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0)
```

# Объединение данных

Соберем бинарные и количественные в одну таблицу

```python
data = pd.concat((data_numerical, data_nonbinary, data[binary_columns]), axis = 1)
print(data.shape)
```

# X и y

```python
X = data.drop(('type'), axis = 1) # выбрасываем столбец 'action'
y = data['type']
feature_names = X.columns
```

Разбивает набор на тренировочный и тестовый. Тестовый набор составляет 25% от общего

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 11)

N_train = X_train.shape 
N_test = X_test.shape 

print(N_train, N_test)
```

```python
print("train = ", N_train[0] / data.shape[0])
print("test = ", N_test[0] / data.shape[0])
```

# $k$NN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train, y_train)
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

Данные удачные при небольшом колечстве соседей (ошибок 0)

```python
knn = KNeighborsClassifier(n_neighbors = 1000)
knn.fit(X_train, y_train)
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```
